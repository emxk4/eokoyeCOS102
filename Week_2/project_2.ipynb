```python
import numpy
def solve_equations():
    # Quadratic Section
    print("Quadratic Equation Calculator")
    # Equation: Ax^2 + Bx + C = 0
    a = float(input("Enter value for A: "))
    b = float(input("Enter value for B: "))
    c = float(input("Enter value for C: "))
    
    # roots() takes a list of coefficients [A, B, C]
    quad_roots = numpy.roots([a, b, c])
    print(f"The roots are: {quad_roots}\n")

    #  Cubic Section 
    print("Cubic Equation Calculator")
    # Equation: Ax^3 + Bx^2 + Cx + D = 0
    a = float(input("Enter value for A: "))
    b = float(input("Enter value for B: "))
    c = float(input("Enter value for C: "))
    d = float(input("Enter value for D: "))
    
    cubic_roots = numpy.roots([a, b, c, d])
    print(f"The roots are: {cubic_roots}\n")

    #  Quartic Section 
    print("Quartic Equation Calculator")
    # Equation: Ax^4 + Bx^3 + Cx^2 + Dx + E = 0
    a = float(input("Enter value for A: "))
    b = float(input("Enter value for B: "))
    c = float(input("Enter value for C: "))
    d = float(input("Enter value for D: "))
    e = float(input("Enter value for E: "))
    
    quartic_roots = numpy.roots([a, b, c, d, e])
    print(f"The roots are: {quartic_roots}")

solve_equations()
```
