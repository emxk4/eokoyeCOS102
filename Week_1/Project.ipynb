```python
print("Welcome to the Compound Interest Calculator")
p = float(input("Enter your principal"))
r = float(input("Enter the rate"))
t = float(input("Enter the time"))
n = float(input("Enter the amount of times interest is compunded per year"))

a = p * (1 + r / n) ** (n * t)
print(f"The compound interest is: {a:.2f}")

print("Welcome to the Simple Interest Calculator")
p = float(input("Enter your principal"))
r = float(input("Enter the rate"))
t = float(input("Enter the time"))

a = p * (1 + (r / 100) * t)
print(f"The simple interest is: {a:.2f}")


print("Welcome to the Annuity Plan Calculator")
p = float(input("Enter your principal"))
r = float(input("Enter the rate"))
t = float(input("Enter the time"))
n = float(input("Enter the amount of times interest is compunded per year"))
pmt = float(input("Enter the periodic payment: "))

a = pmt * (((1 + r / n) ** (n * t) - 1) / (r / n))
print(f"The annuity plan is: {a:.2f}")


          
```
